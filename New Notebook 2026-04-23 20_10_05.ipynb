```python
%sql
CREATE TABLE IF NOT EXISTS employee (
  emp_id INT,
  emp_name STRING,
  department_id INT
);
```

```python
%sql
CREATE TABLE IF NOT EXISTS departments (
  department_id INT,
  department_name STRING
);
```

```python
%sql
 
SELECT e.emp_id, e.emp_name, d.department_name
from employee e
INNER JOIN departments d
ON e.department_id = d.department_id;
```

```python
%sql
SELECT e.emp_id, e.emp_name, d.department_name
from employee e
LEFT JOIN departments d
ON e.department_id = d.department_id;
```

```python
%sql
SELECT e.emp_id, e.emp_name, d.department_name
from employee e
RIGHT JOIN departments d
ON e.department_id = d.department_id;
```

```python
%sql
SELECT e.emp_id, e.emp_name, d.department_name
from employee e
FULL OUTER JOIN departments d
ON e.department_id = d.department_id;
```

```python
%sql
SELECT e.emp_id, e.emp_name, d.department_name
FROM employee e
CROSS JOIN departments d;
```

-- Row_number() - unique, consective, numbering (pagination)
-- rank() - to rank items where ties skip intermediate rank (1,2,2,4)
-- dense_rank() - for ranking without gaps (1,2,2,3)
-- Row_number - Unique rank(no duplicates)
-- Rank() - skips rank() - 1,2,2,4
-- DENSE_RANK() - No skipping - no skipping (1,2,2,3)
-- Row_number() - unique, consective, numbering (pagination) - assigning unique IDs, or picking the top 1 item when ties doesn't matter
 
-- rank() - to rank items where ties skip intermediate rank (1,2,2,4)
-- assign the same rank to ties but leaves a gap in sequence e.g., sports ranking, where tied scores skips the next ranking position
 
-- dense_rank() - for ranking without gaps (1,2,2,3) -
-- assign the same rank to ties without gaps (1,2,2,3)
-- Leaderboards -> where ranking must be contiguous, identifying top N products
 
 
e.g.,
     rn
100  1  1  1
100  2  1  1
80   3  3  2
70   4  4  3
