```python
%sql
SELECT p.post_id, p.post_text, COUNT(r.post_id) AS reactions
FROM workspace.default.facebook_posts p
LEFT JOIN workspace.default.facebook_reactions r
ON p.post_id = r.post_id
GROUP BY p.post_id, p.post_text
ORDER BY reactions DESC;
```
